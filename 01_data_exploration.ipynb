```python
import pandas as pd
import os

# 1. Check current location
current_dir = os.getcwd()
print(f"Jupyter is currently running in: {current_dir}")

# 2. Smart Path Resolution
if current_dir.endswith('notebooks'):
    # Agar hum notebooks folder ke andar hain
    file_path = '../data/processed/master_dataset.csv'
else:
    # Agar hum bahar main project folder mein hain
    file_path = 'data/processed/master_dataset.csv'

print(f"Loading dataset from: {file_path}")

# 3. Load the Data
df = pd.read_csv(file_path)

# 4. Shape Inspection
print(f"\nTotal Rows (Orders + Items): {df.shape[0]}")
print(f"Total Columns (Features): {df.shape[1]}")

# 5. Duplicates Check
duplicates = df.duplicated().sum()
print(f"Total Duplicate Rows: {duplicates}")

# 6. Memory Usage Check
print("\nDataset Info & Memory Usage:")
df.info(memory_usage='deep')
```
